```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read csv files to dataframe

train = pd.read_csv('Train_data.csv')
test = pd.read_csv('Test_data.csv')
train.head()
```

```python
# set test default class data
test_list = [test]
for dataset in test_list:
  dataset['class'] = 0
```

```python
train_list = [train]
for dataset in train_list:
  dataset['class'] = dataset['class'].map( {'normal': 0, 'anomaly': 1} ).astype(int)
```

```python
original_train = train.copy()
full_data = train_list + test_list
```

```python
def get_service_category(name):
  if name.find('ssh') != -1 or name.find('telnet') != -1:
    return 0
  if name.find('ftp') != -1 or name.find('tfpt') != -1:
    return 1
  if name.find('smtp') != -1 or name.find('imap') != -1 or name.find('pop') != -1:
    return 2
  if name.find('http') != -1:
    return 3
  if name.find('systat') != -1 or name.find('netstat') != -1:
    return 4
  if name.find('hostname') != -1 or name.find('domain') != -1:
    return 5
  if name.find('icmp') != -1 or name.find('eco') != -1 or name.find('ecr') != -1:
    return 6
  if name.find('other') != -1:
    return 7
  if name.find('private') != -1:
    return 8
  # undefined
  return 9

def get_flag_category(name):
  if name.find('RSTOS0') != -1 or name.find('SH') != -1:
    return 4
  if name.find('S0') != -1 or name.find('REJ') != -1:
    return 0
  if name.find('S1') != -1 or name.find('SF') != -1 or name.find('OTH') != -1:
    return 1
  if name.find('S2') != -1 or name.find('RSTO') != -1:
    return 2
  if name.find('S3') != -1 or name.find('RSTR') != -1:
    return 3
  # undefined
  return 5
```

```python
for dataset in full_data:
  dataset['protocol_type'] = dataset['protocol_type'].map( {'tcp': 0, 'udp': 1, 'icmp': 2} ).astype(int)
  dataset['service'] = dataset['service'].apply(get_service_category)
  dataset['flag'] = dataset['flag'].apply(get_flag_category)
  # dataset['class'] = dataset['class'].map( {'normal': 0, 'anomaly': 1} ).astype(int)
```
